### Cek korelasi

### Membangun Model

Dari hasil visualisasi diatas bahwa fitue/kolom enginesize memiliki korelasi yang tinggi terhadap kolom price/variable dependant sehingga kita mengambil fitur/kolom enginesize untuk di training 

- Independent variable(x) adalah enginesize
- Dependent variable(y) adalah price

```python
# Mengubah bentuk matriks dengan reshape()
x = dataset['enginesize'].values.reshape(-1,1)
y = dataset['price'].values.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2, random_state=100)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
# Bagian ini cuma ngecek nilai koefisien dan intersep 
print(regressor.coef_)
print(regressor.intercept_)
```

```python
# Cek score model
regressor.score(X_test, y_test)
```

```python
# Tampil nilai X_test
print(X_test)
```

```python
# Prediksi data test
y_prediksi = regressor.predict(X_test)

plt.scatter(X_test, y_test)
plt.plot(X_test, y_prediksi, c='r')
plt.ylabel('Price')
plt.title('Plot enginesize vs Price')
```

```python
# prediksi harga mobil dengan nilai enginesize tertentu
print("nilai prediksi dengan engine size 100 : ", regressor.predict([[100]])) 
print("nilai prediksi dengan engine size 150 : ", regressor.predict([[150]])) 
print("nilai prediksi dengan engine size 200 : ", regressor.predict([[200]])) 
```
